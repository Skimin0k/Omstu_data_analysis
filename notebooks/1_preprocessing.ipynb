### Предобработка набора данных "крестики-нолики"

```python
import pandas as pd
```

```python
data_ttt = pd.read_csv('../data/tic_tac_toe.data', delimiter = ',')
```

```python
data_ttt.head(10)
```

Перекодируем значения набора данных следующим образом. Пустота b - 0, Крестик x - 1, Нолик - 2. Изменим имя колонки 10 на result, а значения: negative - 0, positive - 1.

```python
data_ttt.rename(columns = {'10':'result'})

for column_name in data_ttt:
    data_ttt[column_name] = data_ttt[column_name].map({'b': 0, 'x' : 1, 'o': 2, 'negative': 0, 'positive': 1})
```

```python
data_ttt.isnull().sum()
```

В наборе данных отсутствую пропуски

```python
data_ttt.head(10)
```

```python
# вывод данных в файл
data_ttt.to_csv('tic_tac_toe_preprocessed.csv')
```
